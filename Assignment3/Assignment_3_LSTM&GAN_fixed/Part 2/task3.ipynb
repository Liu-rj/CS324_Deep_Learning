```python
import torch
import matplotlib.pyplot as plt
from my_gan import Generator
```

```python
image1 = torch.randn(1, 100)
image2 = torch.randn(1, 100)
g = Generator(100)
g.load_state_dict(torch.load('models/mnist_generator.pt'))
g.eval()

fig, axs = plt.subplots(nrows=1, ncols=9, figsize=(9, 6), subplot_kw={'xticks': [], 'yticks': []})

for it, ax in enumerate(axs):
    img = ((9 - it) * image2 + it * image1) / 9
    ax.imshow(g(img).detach().numpy().reshape(28, 28), cmap='gray')

plt.tight_layout()
plt.subplots_adjust(hspace=0, wspace=0)
plt.show()
```
