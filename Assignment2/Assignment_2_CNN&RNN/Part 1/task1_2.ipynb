```python
import pytorch_train_mlp as pt_train
import train_mlp_numpy as np_train
import argparse
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons, make_circles, make_blobs
```

```python
DNN_HIDDEN_UNITS_DEFAULT = "20"
LEARNING_RATE_DEFAULT = 1e-2
MAX_EPOCHS_DEFAULT = 200
EVAL_FREQ_DEFAULT = 10
parser = argparse.ArgumentParser()
parser.add_argument(
    "--dnn_hidden_units",
    type=str,
    default=DNN_HIDDEN_UNITS_DEFAULT,
    help="Comma separated list of number of units in each hidden layer",
)
parser.add_argument(
    "--learning_rate",
    type=float,
    default=LEARNING_RATE_DEFAULT,
    help="Learning rate",
)
parser.add_argument(
    "--max_steps",
    type=int,
    default=MAX_EPOCHS_DEFAULT,
    help="Number of epochs to run trainer.",
)
parser.add_argument(
    "--eval_freq",
    type=int,
    default=EVAL_FREQ_DEFAULT,
    help="Frequency of evaluation on the test set",
)
parser.add_argument(
    "--dataset", type=str, default="make_moon", help="Dataset type."
)
parser.add_argument('--optimizer', type=str, default='SGD',
                        help='Optimizer to conduct gradient update')
FLAGS, unparsed = parser.parse_known_args()
```

```python
def draw_eval(epochs, train_acc, test_acc, train_loss, test_loss):
    plt.figure(1, figsize=(20, 5))
    plt.subplot(121)
    plt.plot(epochs, train_acc, 'r', label='Train Acc')
    plt.plot(epochs, test_acc, 'b', label='Test Acc')
    plt.legend()
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(FLAGS.dataset + " dataset SGD training")

    plt.subplot(122)
    plt.plot(epochs, train_loss, 'r', label='Train Acc')
    plt.plot(epochs, test_loss, 'b', label='Test Acc')
    plt.legend()
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(FLAGS.dataset + " dataset SGD training")
    plt.show()
```

```python
FLAGS.dataset = 'make_moon'
dataset = make_moons(n_samples=1000, shuffle=True)
epochs, train_acc, train_loss, test_acc, test_loss = pt_train.train(FLAGS, dataset, 'cpu')
draw_eval(epochs, train_acc, test_acc, train_loss, test_loss)
```

```python
epochs, train_acc, train_loss, test_acc, test_loss = np_train.train(FLAGS, dataset)
draw_eval(epochs, train_acc, test_acc, train_loss, test_loss)
```

```python
FLAGS.dataset = 'make_circles'
dataset = make_circles(n_samples=1000, shuffle=True)
epochs, train_acc, train_loss, test_acc, test_loss = pt_train.train(FLAGS, dataset, 'cpu')
draw_eval(epochs, train_acc, test_acc, train_loss, test_loss)
```

```python
epochs, train_acc, train_loss, test_acc, test_loss = np_train.train(FLAGS, dataset)
draw_eval(epochs, train_acc, test_acc, train_loss, test_loss)
```

```python
FLAGS.dataset = 'make_classification'
dataset = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=1, cluster_std=2)
epochs, train_acc, train_loss, test_acc, test_loss = pt_train.train(FLAGS, dataset, 'cpu')
draw_eval(epochs, train_acc, test_acc, train_loss, test_loss)
```

```python
epochs, train_acc, train_loss, test_acc, test_loss = np_train.train(FLAGS, dataset)
draw_eval(epochs, train_acc, test_acc, train_loss, test_loss)
```
